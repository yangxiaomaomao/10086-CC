```python
import os
import matplotlib.pyplot as plt
```

```python
fname = "bbr-cwnd.txt"
cwnd_list = []
time_list = []
def get_cwnd(s):
    #print(s)
    for item in s.split(" "):
        if "cwnd" in item:
            return int(item.split(":")[1])
        
with open(fname,"r") as f:
    lines = f.readlines()
    start_time = float(lines[0].split(" ")[0])
    #print(start_time)
    for line in lines:
        if "." not in line.split(" ")[0]:
            continue
        #print(line.split(" ")[0])
        time_list.append(float(line.split(" ")[0]) - start_time)
        cwnd_list.append(get_cwnd(line))
        #print(cwnd_list)
```

```python
plt.style.use('ggplot')
plt.plot(time_list,cwnd_list)
plt.title("cwnd-time")
plt.xticks(range(0,int(max(time_list)),int(max(time_list)/10)))
plt.yticks(range(0,int(max(cwnd_list)),int(max(cwnd_list)/10)))
plt.xlabel("time/s")
plt.ylabel("cwnd")
plt.savefig("bbr.pdf")
```
